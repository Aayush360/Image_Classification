### Image Histogram

```python
img_3=skimage.io.imread('images/landscape.jpg')
```

```python
plt.imshow(img_3)
```

```python
img_3.shape
```

```python
img_3=skimage.io.imread('images/landscape.jpg',as_gray=True)
```

```python
plt.imshow(img_3,cmap='gray')
```

```python
img_3.shape
```

```python
flatten_image = img_3.ravel()
```

```python
# plotting the histogram
```

```python
plt.hist(flatten_image,bins=255)
plt.show()
```

#### histogram equalization


```python
## Histogram equalization is a method to process images in order to 
# adjust the contrast of an image by modifying the intensity distribution of the histogram. 
```

```python
from skimage.exposure import equalize_hist
```

```python
equalized_img= equalize_hist(img_3,nbins=256)
```

```python
plt.imshow(equalized_img,cmap='gray')
```

```python
skimage.io.imsave('landscape.jpg',equalized_img)
```

### Image Resize

```python
img_4=skimage.io.imread('images/coffee.jpg')
```
